```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
folder_list = ['F0', 'F1', 'F2']
file_list = ['bond_dim.dat', 'OSEE.dat', 'purity.dat', 'trace.dat', 'total_number_dn.dat', 'total_number_up.dat']
dataframes1 = []
for fn in file_list:
    df1 = pd.read_csv(folder_list[2]+ '/' + fn, sep = '\t', header = None)
    dataframes1.append(df1)


#display(df1)
tot_num = np.array(dataframes1[-1].iloc[:, 2] + dataframes1[-2].iloc[:, 2])
trace = np.array(dataframes1[3].iloc[:, 2])
linear_entropy = 1 - np.array(dataframes1[2].iloc[:, 2])
ent_entropy = np.array(dataframes1[1].iloc[:, 2])
```

```python
df2 = pd.read_csv(r'GSKL_Exact_gamma_N4_U5_densityxt.csv')

Nup = np.zeros((len(df2.iloc[:, 0])))
Ndn = np.zeros((len(df2.iloc[:, 0])))

for j in range(4):
    Nup += np.array(df2.iloc[:, 1+2*j])
    Ndn += np.array(df2.iloc[:, 1+2*j+1])


#plt.plot(Nup)
#plt.plot(Ndn)
fig = plt.figure(figsize = (12, 8))

plt.plot(Nup+Ndn, '-o', label = r'Exact Diagonalization')
plt.plot(tot_num[:50], '-o', label = r'Matrix Product States (BD = 256)')

plt.xlabel('Tempo', fontsize = 20)
plt.ylabel(r'$N_{tot} = \langle N_{\uparrow} + N_{\downarrow}\rangle$', fontsize = 20)
plt.tick_params(labelsize = 15)

plt.grid(visible = True)

plt.legend(loc = 'upper left', fontsize = 15)

plt.show
```
